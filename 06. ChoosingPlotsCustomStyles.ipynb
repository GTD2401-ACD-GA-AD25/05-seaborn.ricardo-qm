```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
spotify_filepath = "https://media.githubusercontent.com/media/GTD2401-ACD-GA-AD25/ACD.datasets.kaggle/refs/heads/main/spotify.csv"

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
# Change the style of the figure to the "dark" theme
sns.set_style("dark")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
# Change the style of the figure to the "dark" theme
sns.set_style("whitegrid")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
# Change the style of the figure to the "dark" theme
sns.set_style("darkgrid")

# Line chart 
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
```

```python
# Personalizar el contexto para presentaciones
sns.set_context("talk", font_scale=1.2)

plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
plt.title("Reproducciones de Spotify")
plt.xlabel("Fecha")
plt.ylabel("Reproducciones")
plt.show()
```
